```python
#W9A3
n=input()
nums=[]
lst=[]
for i in n:
    nums.append(int(i))


def can_min(nums):
    if len(nums)==0:
        lst[-2],lst[-1]=lst[-1],lst[-2]
        return []


    if nums[0]==min(nums):
        lst.append(nums.pop(0))
        return can_min(nums)
    else:
        minval=nums[0]
        i_min=0
        for i in range(len(nums)):
            if nums[i]<minval:
                i_min=i
                minval=nums[i]
        for i in range(len(nums)):
            if nums[i]>minval:
                nums[i],nums[i_min]=nums[i_min],nums[i]
                break
        return nums


ans=lst+can_min(nums)
somin=""
for i in ans:
    somin+=str(i)
print(int(somin))
```

```python
#W9A4

q = int(input())

for _ in range(q):

    a = input()
    b = input()

    n = len(a)
    m = len(b)


    dp = [[False] * (m + 1) for _ in range(n + 1)]
    

    dp[0][0] = True


    for i in range(1, n + 1):
        if a[i-1].islower():
            dp[i][0] = dp[i-1][0]
        else:
            dp[i][0] = False


    for i in range(1, n + 1):
        for j in range(1, m + 1):
            char_a = a[i-1]
            char_b = b[j-1]

            if char_a.isupper():

                if char_a == char_b:
                    dp[i][j] = dp[i-1][j-1]
                else:
                    dp[i][j] = False
            else:

                delete_opt = dp[i-1][j]
                
                match_opt = False
                if char_a.upper() == char_b:
                    match_opt = dp[i-1][j-1]
                
                dp[i][j] = delete_opt or match_opt

    if dp[n][m]:
        print("YES")
    else:
        print("NO")
```

```python
#W9A7
def canplaceflower(flower,k):
    
    flower = [0] + flower + [0]
    n = len(flower)
    for i in range(1,n-1):
        if flower[i] == 0 and flower[i+1] == 0 and flower[i-1] == 0:
            k -= 1
            flower[i] = 1
    return k <= 0
flower = list(map(int,input().split()))
k = int(input())
print('True' if canplaceflower(flower,k) else "False")
```

```python
#W9A8
path = {'R':[0,1],'L':[0,-1],'U':[-1,0],'D':[1,0]}
a={(0,0)}
def is_path_crossing(moves):
    i,j = 0,0
    for ch in moves:
        i1 = i + path[ch][0]
        j1 = j + path[ch][1]
        if (i1,j1) in a:
            return True
        a.add((i1,j1))
        i,j = i1,j1
    return False
moves = input()
if is_path_crossing(moves):
    print('True')
else:
    print('False')
```

```python
#W9A9
def Median(nums):
    a,b,c,d,e = nums
    if a > b:
        a,b=b,a
    if c>d:
        c,d=d,c
    if a>c:
        a,c=c,a
        b,d=d,b
    if e < b:
        return b
    if e > c:
        return c
    return e
nums = list(map(int,input().split()))
print(Median(nums))
```

```python
#W9A10
def findRadius(houses,heaters):
    houses.sort()
    heaters.sort()
    i = 0
    r = 0
    for h in houses:
        while i < len(heaters) -1 and abs(heaters[i+1]-h) <= abs(heaters[i]-h):
            i+=1
        r = max(r,abs(heaters[i] - h))
    return r
houses = list(map(int,input().split()))
heaters = list(map(int,input().split()))
print(findRadius(houses,heaters))
```
