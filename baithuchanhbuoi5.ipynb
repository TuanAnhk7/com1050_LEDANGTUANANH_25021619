```python
#W5A1
def Max_Of_Two( a,b):
    if a>b:
        return(a)
    return(b)
m,n=map(int,input().split())
print(Max_Of_Two(m,n))
```

```python
#W5A2
def swap(a,b):
    a,b=b,a
    return a,b
a,b=map(int,input().split())
m,n=swap(a,b)
print(m,n)
```

```python
#W5A3
def is_Prime(n):
    if n==0==1:
        return False
    for i in range(2,int(n**0.5)+1):
        if n%i==0:
           return False
    return True
n=int(input())
print(is_Prime(n))
```

```python
#W5A4
def perfect(n):
    sum=0
    for i in range(1,n):
        if n%i==0:
            sum+=i
    if sum==n:
        return True
    else:
        return False
n=int(input())
print(perfect(n))
```

```python
#W5A5
def the_posision_of_k(n):
    cnt=0
    ans=False
    for i in n:
        cnt+=1
        if i==k:
            ans=True
            break
    if ans:
        return cnt
    else:
        return None
n=list(map(int,input().split()))
k=int(input())
print(the_posision_of_k(n))
```

```python
#W5A6
def factorial(n):
    S=1
    for i in range(2,n+1):
        S*=i
    return S
n=int(input())
print(factorial(n))
```

```python
#W5A7
a,o,b=map(str,input())
a=float(a)
b=float(b)
def casio(a,o,b):
    if o=="+":
        return a+b
    elif o=="-":
        return a-b
    elif o=="/":
        return a/b
    else:
        return a*b
print(f"{casio(a,o,b):.2f}")
```

```python
#W5A8
def bit(n):
    bit=""
    while n>0:
        bit=bit+str(n%2)
        n//=2
    return bit
def distance_hamming(a,b):
    cnt=0
    m,n=bit(a),bit(b)
    if len(m)>len(n):
        while len(m)!=len(n):
            n+="0"
    else:
        while len(n)!=len(m):
            m+="0"
    m=m[::-1]
    n=n[::-1]
    for i in range(len(m)):
        if m[i]!=n[i]:
            cnt+=1
    return cnt
a,b=map(int,input().split())
print(distance_hamming(a,b))
```

```python
#W5A9
def sum_gigit(n):
    sum=0
    while n>0:
        sum+=n%10
        n//=10
    return sum
n=int(input())
print(sum_gigit(n))
```

```python
#W5A10
def isomorphism(a,b):
    ans=True
    for i in range(len(a)):
        for j in range(len(a)):
            if a[i]==a[j]:
                if b[i]!=b[j]:
                    ans=False
    return ans
def is_isomorphism():
    if isomorphism(a,b) and isomorphism(b,a):
        return True
    return False
a,b=map(str,input().split())
print(is_isomorphism())
```
