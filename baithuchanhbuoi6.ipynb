```python
#W6A1
n=list(map(int,input().split()))
lst=[]
def remove_duplicates(n):
    for i in n:
        if i not in lst:
            lst.append(i)
    return(lst)
print(remove_duplicates(n))
```

```python
#W6A2
n=list(map(int,input().split()))
def cumulative_sum(n):
    for i in range(1,len(n)):
        n[i]=n[i]+n[i-1]
    return(n)
print(cumulative_sum(n))
```

```python
#WA6A5
n=list(map(int,input().split()))
def is_number(n):
    dic={}
    pos=0
    nega=0
    zero=0
    for i in n:
        if i==0:
            zero+=1
        elif i>0:
            pos+=1
        else:
            nega+=1
    dic['positives']=pos
    dic['negatives']=nega
    dic['zero']=zero
    return dic
print(is_number(n))
```

```python
#W6A10
lst=list(map(int,input().split()))
k=int(input())
lst1=[]
def sum_equal_k(lst):
    for i in range(len(lst)):
        for j in range(i+1,len(lst)):
            if lst[i]+lst[j]==k:
                if (lst[i],lst[j]) not in lst1:
                    lst1.append((lst[i],lst[j]))
    lst1.sort()
    return lst1
print(sum_equal_k(lst))
```

```python
#W6A12
lst=list(map(int,input().split()))
dict={}
lst.sort()
for i in lst:
    if i not in dict:
        dict[i]=1
    else:
        dict[i]+=1
max=0
frequency=0
for key,value in dict.items():
    if value>max:
        max=value
        frequency=key
print(frequency)
```

```python
#W6A16
n = int(input())
m = int(input())
matrix = []

# Nhập dữ liệu
for i in range(n):
    row_data = []
    for j in range(m):
        val = int(input())
        row_data.append(val)
    matrix.append(row_data)

# In kết quả
for i in range(n):
    for j in range(m):
        print(f"{matrix[i][j]:>4}", end="")
    print()
```

```python
#W6A17
n=int(input())
matrix=[]
main=[]
extra=[]
for i in range(n):
    row=[]
    for j in range(n):
        val=int(input())
        row.append(val)
    matrix.append(row)
for i in range(n):
    for j in range(n):
        if i==j:
            main.append(matrix[i][j])
        if i+j==n-1:
            extra.append(matrix[i][j])
print(main)
print(extra)
```
