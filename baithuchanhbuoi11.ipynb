```python
#W10A3
try:
    n=int(input())
    lst=[n]
    n=str(n)
    s=0
    while s!=1:
        s=0
        for i in n:
            s+=(int(i))**2
        if s in lst:
            break
        lst.append(s)
        n=str(s)
    if s==1:
        print("YES")
    else:
        print("NO")
except:
    print("Loi")

```

```python
#W10A11
import re

try:
    n_input = input()
    if not n_input: exit()
    n = int(n_input)
    
    lst = []
    for _ in range(n):
        s = input().strip() 
        lst.append(s)

    for s in lst:

        if "*+" in s or "++" in s or "?+" in s:
            print(False)
            continue 

        try:
            re.compile(s)
            print(True)
        except re.error:
            print(False)
            
except:
    pass
```
