```python
#W8A1
class cylinder:
    def __init__(self,r,h):
        self.r=r
        self.h=h
    def surface_area(self):
        self.S=self.r*2*3.14*self.h+self.r**2*3.14*2
        return self.S
    def volume(self):
        self.V=self.r**2*3.14*self.h
        return self.V
r,h=map(float,input().split())
ans=cylinder(r,h)
print(f'{ans.surface_area():.2f} {ans.volume():.2f}')
```

```python
#6
class Rectangle:
    def __init__(self, w, h):
        self.w = w
        self.h = h

    def area(self):
        return self.w * self.h

    def perimeter(self):
        return 2 * (self.w + self.h)

    def scale(self, k):
        self.w *= k
        self.h *= k

try:
    line = input().strip()
    if line:
        values = list(map(float, line.split()))
        if len(values) == 3:
            w, h, k = values
            rect = Rectangle(w, h)
            rect.scale(k)

            print(f"{rect.area():.2f} {rect.perimeter():.2f}")
except:
    pass
```

```python
#7
import math

class Fraction:
    def __init__(self, num, den):
        if den == 0:
            raise ValueError
        common = math.gcd(num, den)
        self.num = num // common
        self.den = den // common
        if self.den < 0:
            self.num = -self.num
            self.den = -self.den

    def __str__(self):
        return f"{self.num}/{self.den}"

    def add(self, other):
        return Fraction(self.num * other.den + other.num * self.den, self.den * other.den)

    def sub(self, other):
        return Fraction(self.num * other.den - other.num * self.den, self.den * other.den)

    def mul(self, other):
        return Fraction(self.num * other.num, self.den * other.den)

    def div(self, other):
        return Fraction(self.num * other.den, self.den * other.num)
try:
    line = input().split()
    if len(line) == 5:
        a, b, op, c, d = int(line[0]), int(line[1]), line[2], int(line[3]), int(line[4])
        f1 = Fraction(a, b)
        f2 = Fraction(c, d)

        if op == '+':
            print(f1.add(f2))
        elif op == '-':
            print(f1.sub(f2))
        elif op == '*':
            print(f1.mul(f2))
        elif op == '/':
            print(f1.div(f2))
except:
    pass
```

```python
#8
class BankAccount:
    def __init__(self, owner, balance):
        self.owner = owner
        self.balance = float(balance)

    def deposit(self, x):
        self.balance += x

    def withdraw(self, x):
        if x <= self.balance:
            self.balance -= x

    def __str__(self):
        return f"{self.balance:.2f}"

try:
    line1 = input().split()
    if len(line1) >= 2:
        owner = line1[0]
        initial_balance = float(line1[1])
        account = BankAccount(owner, initial_balance)

        n = int(input())
        for _ in range(n):
            transaction = input().split()
            cmd = transaction[0]
            amount = float(transaction[1])
            if cmd == "DEPOSIT":
                account.deposit(amount)
            elif cmd == "WITHDRAW":
                account.withdraw(amount)

        print(account)
except:
    pass
```

```python
#9
class Student:
    def __init__(self, name):
        self.name = name
        self.scores = []

    def add_score(self, subject, score):
        self.scores.append(score)

    def avg(self):
        if not self.scores:
            return 0.0
        return sum(self.scores) / len(self.scores)

    def rank(self):
        average = self.avg()
        if average >= 8:
            return "Excellent"
        elif average >= 6.5:
            return "Good"
        elif average >= 5:
            return "Average"
        else:
            return "Poor"

    def __str__(self):
        return f"{self.name} {self.avg():.2f} {self.rank()}"
try:
    name = input().strip()
    n_str = input().strip()
    if name and n_str:
        n = int(n_str)
        student = Student(name)

        for _ in range(n):
            line = input().split()
            if len(line) >= 2:
                subject = line[0]
                score = float(line[1])
                student.add_score(subject, score)

        print(student)
except EOFError:
    pass
except Exception:
    pass
```

```python
#10
class Book:
    def __init__(self, title, author, year):
        self.title = title
        self.author = author
        self.year = int(year)

class Library:
    def __init__(self):
        self.books = []

    def add(self, book):
        self.books.append(book)

    def count_by_author(self, name):
        count = 0
        for book in self.books:
            if book.author == name:
                count += 1
        return count

    def find_by_year(self, y):
        count = 0
        for book in self.books:
            if book.year == int(y):
                count += 1
        return count
try:
    n_str = input().strip()
    if n_str:
        n = int(n_str)
        library = Library()

        for _ in range(n):
            line = input().strip()
            if not line:
                continue

            parts = line.split(' ', 1)
            cmd = parts[0]

            if cmd == "ADD":
                details = parts[1].split(';')
                if len(details) == 3:
                    book = Book(details[0], details[1], details[2])
                    library.add(book)
            elif cmd == "COUNT":
                author_name = parts[1]
                print(library.count_by_author(author_name))
            elif cmd == "COUNTYEAR":
                year_val = parts[1]
                print(library.find_by_year(year_val))
except EOFError:
    pass
except Exception:
    pass
```

```python

#11
import math

class Vector3D:
    def __init__(self, x, y, z):
        self.x = x
        self.y = y
        self.z = z

    def __sub__(self, other):
        return Vector3D(self.x - other.x, self.y - other.y, self.z - other.z)

    def cross(self, other):
        return Vector3D(
            self.y * other.z - self.z * other.y,
            self.z * other.x - self.x * other.z,
            self.x * other.y - self.y * other.x
        )

    def dot(self, other):
        return self.x * other.x + self.y * other.y + self.z * other.z

    def magnitude(self):
        return math.sqrt(self.x**2 + self.y**2 + self.z**2)
def get_angle(n1, n2):
    dot_product = n1.dot(n2)
    m1 = n1.magnitude()
    m2 = n2.magnitude()
    cos_phi = dot_product / (m1 * m2)
    cos_phi = max(-1.0, min(1.0, cos_phi))
    phi_rad = math.acos(cos_phi)
    return math.degrees(phi_rad)

try:
    points = []
    for _ in range(4):
        line = input().split()
        if line:
            points.append(Vector3D(float(line[0]), float(line[1]), float(line[2])))

    if len(points) == 4:
        A, B, C, D = points

        vAB = B - A
        vBC = C - B
        vCD = D - C

        n1 = vAB.cross(vBC)
        n2 = vBC.cross(vCD)

        phi = get_angle(n1, n2)
        print(f"{phi:.2f}")
except:
    pass
```

```python
#12
class ComplexNumber:
    def __init__(self, realPart, imaginaryPart):
        self.realPart = float(realPart)
        self.imaginaryPart = float(imaginaryPart)

    def multiply(self, other):
        new_real = self.realPart * other.realPart - self.imaginaryPart * other.imaginaryPart
        new_imaginary = self.realPart * other.imaginaryPart + self.imaginaryPart * other.realPart
        return ComplexNumber(new_real, new_imaginary)

    def __str__(self):
        return f"{int(self.realPart)} {int(self.imaginaryPart)}"

try:
    line1 = input().split()
    line2 = input().split()

    if len(line1) == 2 and len(line2) == 2:
        c1 = ComplexNumber(line1[0], line1[1])
        c2 = ComplexNumber(line2[0], line2[1])

        result = c1.multiply(c2)
        print(result)
except:
    pass
```

```python
#13
import math

class Triangle:
    def __init__(self, e1, e2, e3):
        self.e1 = e1
        self.e2 = e2
        self.e3 = e3

    def is_valid(self):
        return (self.e1 + self.e2 > self.e3) and (self.e1 + self.e3 > self.e2) and (self.e2 + self.e3 > self.e1)

    def getArea(self):
        if not self.is_valid():
            return -1
        p = (self.e1 + self.e2 + self.e3) / 2
        area = math.sqrt(p * (p - self.e1) * (p - self.e2) * (p - self.e3))
        return area

try:
    line = input().split()
    if len(line) == 3:
        a, b, c = map(int, line)
        tri = Triangle(a, b, c)
        result = tri.getArea()

        if result == -1:
            print("invalid")
        else:
            if result == int(result):
                print(int(result))
            else:
                print(f"{result:.2f}")
    else:
        print("invalid")
except:
    print("invalid")
     
```
