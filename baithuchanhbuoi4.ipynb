```python
#14  
GCD = 1
a, b = map(int, input().split())
A = min(a,b)
B = max(a,b)
i = 1
while i*i <= A:
    if A % i == 0:
        if B % i == 0:
            GCD = max(GCD, i)
        if B % (A // i) == 0:
            GCD = max(GCD, A // i)
    i+=1

print(GCD)
```
