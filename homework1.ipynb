```python
#bai6
a,b,c=map(float,input("Nhập lần lượt độ dài 3 cạnh của tam giác: ").split())
if a+b>c and a+c>b and b+c>a:
    TB=(a+b+c)/2
    s = s = (TB*(TB-a)*(TB-b)*(TB-c))**0.5
    print("Diện tích của tam giác là",f"{s:.1f}")
else:
    print("3 cạnh đó không tạo thành một tam giác")
```
